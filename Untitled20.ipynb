```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('시도별 전출입 인구수.xlsx',header=0) #교재 코드 수정(fillna 밖으로 뺴야함)
```

```python
print(df.head())
```

```python
df=df.fillna(method='ffill')
```

```python
print(df.head())
```

```python
mask=(df['전출지별']=='서울특별시') & (df['전입지별'] != '서울특별시' )
```

```python
df_seoul=df[mask]
```

```python
print(df_seoul)
```

```python
df_seoul=df_seoul.drop(['전출지별'],axis=1)  #axis=1(열을 따라 작동)
```

```python
df_seoul.rename({'전입지별':'전입지'},axis=1,inplace=True)
```

```python
df_seoul.set_index('전입지',inplace=True)
```

```python
print(df_seoul)
```

```python
sr_one=df_seoul.loc['경기도']
plt.plot(sr_one.index,sr_one.values) #plt.plot(x,y) ==plt.plot(sr_one)
plt.title('서울->경기 지역 이동')
plt.xlabel('기간')
plt.ylabel('이동 인구수')
plt.show()
```

#그래프 꾸미기

```python
sr_one =df_seoul.loc['경기도']
plt.figure(figsize=(14,5))  #그래프 크기
plt.style.use('ggplot') #그래프 스타일 서식 지정
plt.xticks(size=10,rotation='vertical') #x축 눈금 라벨 회전
x=sr_one.index
y=sr_one.values
plt.plot(x,y,marker='o',markersize=10)
plt.title('movement')
plt.xlabel('day')
plt.ylabel('# of people')
plt.legend(labels=['seoul->gyeongi'],loc='best',fontsize=15) #labels = (범례 이름) , loc = (위치 지정)
plt.ylim(50000,800000) #y축 크기 지정
plt.annotate('',xy=(20,620000),xytext=(2.290000),arrowprops=dict(arrowstyle='->',color='skyblue',lw=5)) #https://cocoon1787.tistory.com/29 (cannot unpack non-iterable float object 에러 해결)
plt.annotate('',(47,450000),(30.580000),arrowprops=dict(arrowstyle='->',color='olive',lw=5))
plt.annotate('increment of people',(10,550000),rotation=25,va='baseline',ha='center',fontsize=15)
plt.annotate('decrement of people',(40,560000),rotation=11,va='baseline',ha='center',fontsize=15)
plt.show()
```

```python
#한번에 그래프 여러 개 그리기
```

```python
fig=plt.figure(figsize=(10,10))
ax1=fig.add_subplot(2,1,1) #(행의 크기,열의 크기,서브플로 순서)
ax2=fig.add_subplot(2,1,2)
ax1.plot(sr_one,marker='o',markersize=10)
ax2.plot(sr_one,marker='o',markerfacecolor='green',markersize=10,color='olive',linewidth=2,label='seoul->gyeongi')
ax2.legend(loc='best')
ax1.set_ylim(50000,800000)
ax2.set_ylim(50000,800000)
ax1.set_xticklabels(sr_one.index,rotation=75)
ax2.set_xticklabels(sr_one.index,rotation=75)
ax2.tick_params(axis="x",labelsize=10)
ax2.tick_params(axis="y",labelsize=10)
plt.show()
```

#면적 그래프

```python
t=df_seoul.loc[['충청남도','경상북도','강원도','전라남도'],col_years]
t=t.transpose()
t.index=t.index.map(int) #잘 모르겠음
ax=t.plot(kind='area',stacked=False,alpha=0.2,figsize=(20,10)) #stacked=True/False
ax.legend(loc='best')
ax.set_title('movement',size=30)
ax.set_xlabel('length',size=20)
ax.set_ylabel('# of people',size=20)
plt.show()
```

```python
#파이 차트
```

```python
df['count']=1
df_origin=df.groupby('origin').sum() #groupby메소드로 데이터를 숫자로 지정
print(df_origin.head())
```

```python
df_origin.index=['USA','EU','JPN']
```

```python
df_origin['count'].plot(kind='pie',figsize=(10,5),autopct='%1.1f%%',startangle=10,colors=['green','yellow','brown'])
plt.title('Model origin')
plt.axis('equal')
plt.legend(labels=df_origin.index,loc='best')
plt.show() #https://wikidocs.net/92095
```

#Seaborn 라이브러리 https://steadiness-193.tistory.com/201

```python
import seaborn as sns
```

```python
titanic=sns.load_dataset('titanic')
```

```python
print(titanic.info()) #데이터셋의 모든 데이터 타입
```

```python
#회귀선이 있는 산점도 - sns.regplot( x= , y= , data= ,fit_reg=True/False)
```

```python
 #히스토그램/커널 밀도 그래프 - sns.distplot(titanic['fare'],hist/kde=False)
```

```python
#히트맵
```

```python
table=titanic.pivot_table(index=['sex'],columns=['class'],aggfunc='size') #aggfunc=기준,pivot_table=엑셀 테이블 형태로 데이터 수정
```

```python
sns.heatmap(table,annot=True,fmt='d',cmap='YlGnBu',cbar=True)
plt.show()
```

```python
#막대 그래프
```

```python
fig=plt.figure(figsize=(15,5))
a1=fig.add_subplot(1,3,1)
a2=fig.add_subplot(1,3,2)
a3=fig.add_subplot(1,3,3)
sns.barplot(x='sex',y='survived',data=titanic,ax=a1)
sns.barplot(x='sex',y='survived',data=titanic,hue='class',ax=a2) #hue=카테고리 별 지정(범례)
sns.barplot(x='sex',y='survived',data=titanic,hue='class',dodge=False,ax=a3) #dodge=누적
import matplotlib.pyplot as plt
plt.show()
```

```python
#빈도 그래프 -데이터의 개수를 파악하고 싶을 때 : sns.countplot( x= , hue= ,palette= , data=titanic , dodge=False )
```

```python
#특정 위치(ex:대학교) 마커로 표시하기
```

```python
import pandas as pd
```

```python
df=pd.read_excel('서울지역 대학교 위치.xlsx')
print(df)
df.columns=['학교','위도','학교']
print(df)
```

```python
seoul_map=folium.Map(location=[37.55,126.98],zoom_start=12)
for name,lat,lng in zip(df.index,df['위도'],df['경도']):folium.Marker([lat,lng],popup=name).add_to(seoul_map)
seoul_map.save('keke.html')
```
